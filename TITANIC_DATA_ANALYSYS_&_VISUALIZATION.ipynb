```python
import pandas as pd

# Load dataset
df = pd.read_csv('titanic.csv')

# Remove duplicates
df = df.drop_duplicates()

# Fill missing values
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])

# Drop 'deck' column due to many missing values
df = df.drop(columns=['deck'])

# Convert object columns to category
categorical_cols = ['sex', 'embarked', 'class', 'who', 'embark_town', 'alive']
df[categorical_cols] = df[categorical_cols].astype('category')

# Create new feature
df['family_size'] = df['sibsp'] + df['parch'] + 1

# Encode categorical features
df['sex_code'] = df['sex'].cat.codes
df['embarked_code'] = df['embarked'].cat.codes

# View updated data
print(df.head())
```

## Data analysis

### Subtask:
Analyze the distribution of key features and calculate survival rates.


**Reasoning**:
Analyze the distribution of key features and calculate survival rates as requested in the subtask.



`**Missing value and Outliers**`

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
df = pd.read_csv('titanic.csv')

# Display missing values in each column
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Detect outliers using Z-score for numerical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
outlier_counts = {}

for col in numeric_cols:
    col_data = df[col].dropna()
    z_scores = np.abs(stats.zscore(col_data))
    outlier_counts[col] = (z_scores > 3).sum()

print("\nOutliers Detected (Z-score > 3):\n", outlier_counts)
```

```python
# Analyze the distribution of key features
print(df['pclass'].value_counts())
print(df['sex'].value_counts())
print(df['age'].describe())
print(df['embarked'].value_counts())

# Calculate overall survival rate
overall_survival_rate = df['survived'].mean()
print(f"\nOverall Survival Rate: {overall_survival_rate:.2%}")

# Calculate survival rates for different passenger groups
print("\nSurvival Rate by Passenger Class:")
print(df.groupby('pclass')['survived'].mean())

print("\nSurvival Rate by Sex:")
print(df.groupby('sex')['survived'].mean())

print("\nSurvival Rate by Embarked:")
print(df.groupby('embarked')['survived'].mean())

# Analyze survival rate by age bands
def categorize_age(age):
    if age < 16:
        return 'Child'
    elif age < 30:
        return 'Young Adult'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_band'] = df['age'].apply(categorize_age)
print("\nSurvival Rate by Age Band:")
print(df.groupby('age_band')['survived'].mean())

# Explore the relationship between fare and survival (example)
print("\nCorrelation between Fare and Survival:")
print(df[['fare', 'survived']].corr())
```
